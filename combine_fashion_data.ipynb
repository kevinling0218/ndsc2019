```python
import pandas as pd
```

```python
df_1 = pd.read_csv('./data/fashion_data_info_val_competition_old.csv')
df_2 = pd.read_csv('./data/fashion_data_info_val_submission.csv').drop(['Id'], axis = 1)
```

```python
from NLP.util import column_text_to_class, category_json, category_feature_columns
```

```python
df_2
```

```python
map_json = category_json['fashion']
column_map = {}
for column in category_feature_columns['fashion']:
    column_map[column] = {k: v for k, v in map_json[column].items()}
    df_2.loc[:, column] = df_2[column].map(column_map[column])
```

```python
df_combine = pd.concat([df_1, df_2], axis = 1)
```

```python
df_fashion_train = pd.read_csv('./data/fashion_data_info_train_competition_old.csv')
```

```python
df_fashion_train_new = pd.concat([df_combine, df_fashion_train])
```
